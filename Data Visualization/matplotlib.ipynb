```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('titanic')
data.head()
```

```python
data.isnull().sum()
```

```python
plt.boxplot()
```

```python
data.describe()
```

```python
plt.figure(figsize=(19,5))
plt.scatter(x=data['age'],y=data['fare'],c='green',marker='*',s=100,alpha=0.5)
plt.xlabel('age')
plt.ylabel('fare')
plt.title("scatter plot")
plt.show()
```

```python
# Mengelompokkan jumlah penumpang berdasarkan nilai umur
age_counts = data['age'].value_counts().sort_index()
plt.figure(figsize=(10,5))

# Plot grafik garis dengan nilai umur sebagai sumbu X dan jumlah penumpang sebagai sumbu Y
age_counts.plot(kind='line')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers by Age')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(age_counts.index,age_counts.values,color='green',width=1.5)
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers by Age')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(data['age'],bins=20)
```

```python
plt.pie(data['survived'].value_counts(),labels=data['survived'].value_counts().index,autopct="%.2f%%")
plt.show()
```

```python
# Plot box plot dari kolom 'age'
plt.boxplot(data['age'].dropna())  # dropna() untuk menghapus nilai NaN jika ada
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Box Plot of Age')
plt.show()
```
