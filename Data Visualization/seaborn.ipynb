```python
import seaborn as sns
```

```python
data = sns.load_dataset('iris')
data.head()
```

```python
# Scatter Plot
sns.scatterplot(x=data['sepal_length'],y=data['sepal_width'],hue=data['species'],size=data['species'],palette='YlGnBu')
```

```python
# histogram
sns.histplot(x=data['petal_length'],bins=10,kde=True)
```

```python
sns.displot(x=data['petal_length'],bins=10,kde=True)
```

```python
sns.boxplot(x=data['petal_length'])
```

```python
sns.boxplot(x=data['petal_length'],y=data['species'],hue=data['species'])
```

```python
sns.stripplot(x=data['petal_length'],y=data['species'],hue=data['species'])
```

```python
sns.pairplot(data.select_dtypes(['number']))
```

```python
sns.heatmap(data=data.drop(columns=['species']).corr(),annot=True,cmap='YlGnBu')
```
