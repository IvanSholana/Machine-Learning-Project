```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
df = pd.read_csv("./dataset/Churn_Modelling.csv")
df.head(5)
```

```python
df.drop(columns=['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
df.head()
```

### UNSUPERVISED BINNING

```python
labels = ['0-20','21-40','41-60','above 61']
bins = [0,21,41,61,100]

df['Age Bins'] = pd.cut(df['Age'],bins,labels=labels,include_lowest=True)
```

```python
df[['Age', 'Age Bins']].loc[df['Age'] > 61].head(5)
```

```python
df['Age Bins'].value_counts()
```

```python
plt.bar(labels,df['Age Bins'].value_counts())

for i in range(len(labels)):
    plt.text(labels[i],df['Age Bins'].value_counts()[i],df['Age Bins'].value_counts()[i])

plt.title('Age Count')
plt.xlabel('Age Bins')
plt.ylabel('Age Count')

plt.show()
```
