```python
import pandas as pd
```

```python
# Load a Data

df = pd.read_csv("./dataset/pokemon_data.csv",index_col=0,delimiter=',') # Read and create dataframe

df.head() # show top 5 dataframe
# df.tail() show bottom 5 dataframe
```

### Reading Data in Pandas

```python
print(df.columns) # show the header
print(df[['Name','HP']][0:10]) # Show spesific collumn and row
print(df.iloc[1:4][0:4]) # index location (column) (:)
print(df.iloc[1,0]) # baris 1 kolom 0 (,)
```

```python
for x,y in df.iterrows(): # melakukan perulangan terhadap setiap row 
    print(x,y['Name'])
```

```python
print(df.loc[(df['Attack'] > 10) & (df['Type 2'] == 'Grass')].head())
```
