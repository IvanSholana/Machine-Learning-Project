```python
import pandas as pd

# Data contoh
data = {
    'Name': ['Pikachu', 'Bulbasaur', 'Charmander', 'Squirtle'],
    'Type 1': ['Electric', 'Grass', 'Fire', 'Water'],
    'Type 2': [None, 'Poison', None, None],
    'HP': [35, 45, 39, 44],
    'Attack': [55, 49, 52, 48]
}

# Membuat DataFrame
df = pd.DataFrame(data)

# Melakukan pivot
pivot_df = pd.pivot(data=df, index='Name', columns='Type 1', values='Attack')

print(pivot_df)
```

```python
# Melakukan pivot table
pivot_table_df = pd.pivot_table(df, index='Type 1', columns='Type 2', values='Attack', aggfunc='mean')

print(pivot_table_df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate multivariate data
np.random.seed(0)
data = np.random.rand(100, 4)  # 100 data points with 4 features: tinggi, berat, usia, fitur4

# Applying K-Means
kmeans = KMeans(n_clusters=3)  # Defining number of clusters as 3
kmeans.fit(data)
labels = kmeans.labels_  # Get cluster labels
centers = kmeans.cluster_centers_  # Get cluster centers

# Visualization
plt.figure(figsize=(8, 6))

# Plotting data points
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', s=50, alpha=0.5)

# Plotting cluster centers
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.8)

plt.xlabel('Tinggi')
plt.ylabel('Berat')
plt.title('Clustering dengan K-Means pada Data Multivariate dengan 4 Fitur (2 Dimensi)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate multivariate data
np.random.seed(0)
data = np.random.rand(100, 4)  # 100 data points with 4 features: tinggi, berat, usia, fitur4

# Menghitung WCSS untuk berbagai jumlah klaster
wcss_values = []
for k in range(1, 11):  # Coba jumlah klaster dari 1 hingga 10
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data)
    wcss_values.append(kmeans.inertia_)  # inertia_ adalah atribut yang berisi WCSS

# Visualisasi kurva WCSS terhadap jumlah klaster
plt.plot(range(1, 11), wcss_values, marker='o')
plt.xlabel('Jumlah Klaster')
plt.ylabel('WCSS')
plt.title('Elbow Method untuk Penentuan Jumlah Klaster')
plt.show()
```
